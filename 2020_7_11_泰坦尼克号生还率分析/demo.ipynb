```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

# 读取csv数据文件
df = pd.read_csv('./titanic.csv')
# 查看缺失值,发现没有缺失值
df.info()
```

```python
# 统计生还人数
live = 0
# 统计死亡人数
dead = 0
for i in df['Survived']:
    if (i == 1):
        live += 1
    else:
        dead += 1
print("生还人数是{}，未生还人数是{}".format(live,dead))
print("生还人数和未生还人数所占比分别是：{}% , {}%".format(live/(live+dead)*100,dead/(live+dead)*100))
```

```python
# 分析舱位和生还率之间的关系
survived1_1 = df['Pclass_1'][df['Survived'] == 1].value_counts()
survived1_0 = df['Pclass_1'][df['Survived'] == 0].value_counts()
sex_survived = pd.DataFrame({'获救':survived1_1,'死亡':survived1_0})
# 显示的第二栏数据就是当前舱位获救和死亡的人数----->survived1_1[1]和survived1_0[1]
print(sex_survived)
# 下面是2号和3号舱位的获救和死亡人数统计
survived2_1 = df['Pclass_2'][df['Survived'] == 1].value_counts()
survived2_0 = df['Pclass_2'][df['Survived'] == 0].value_counts()
survived3_1 = df['Pclass_3'][df['Survived'] == 1].value_counts()
survived3_0 = df['Pclass_3'][df['Survived'] == 0].value_counts()
# 画柱状图
x_data = ['1_死亡','2_死亡','3_死亡']
x_data1 = ['1_获救','2_获救','3_获救']
# 存放3个舱位死亡数
y_data = [survived1_0[1],survived2_0[1],survived3_0[1]]
# 存放3个舱位获救数
y_data2 = [survived1_1[1],survived2_1[1],survived3_1[1]]
plt.bar(x=x_data,height=y_data,label='死亡人数',color='steelblue',alpha=0.8)
plt.bar(x=x_data1,height=y_data2,label="获救人数",color="indianred",alpha=0.8)
# 设置标题
plt.title("3个舱位死亡和获救人数统计")
# 为两条坐标轴设置名称
plt.xlabel("舱位")
plt.ylabel("人数")
# 显示中文名称
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
# 显示图例
plt.legend()
plt.show()
```
