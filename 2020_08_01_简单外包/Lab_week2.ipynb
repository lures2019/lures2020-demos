## Problem 10

- Plot the clusteirng data using the code on the first slide of week 2 lecture 3.
- How many groups do you observe? Change the number of cluster to vary the clustering result.
- Plot the results with various number of clusters (e.g. 2, 3, 4, etc.)

```python
import numpy as np
import matplotlib.pyplot as plt   ## This is another library for plotting and visualization
import sklearn     
from sklearn.cluster import KMeans
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data
y = iris.target

kmeans8 = KMeans(n_clusters=8)
kmeans8.fit(X)
xc = kmeans8.transform(X)
xc = kmeans8.labels_

plt.scatter(X[:,2],X[:,0],c=xc.astype(np.float))
plt.figure()
plt.scatter(X[:,2],X[:,0],c=y.astype(np.float))
plt.figure()

plt.scatter(X[:,3],X[:,0],c=xc.astype(np.float))
plt.figure()
plt.scatter(X[:,3],X[:,0],c=y.astype(np.float))
```
