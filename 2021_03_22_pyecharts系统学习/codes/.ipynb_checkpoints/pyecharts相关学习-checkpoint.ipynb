```python
import pyecharts.options as ops
from pyecharts.charts import WordCloud

names = [
        '', '', '', '', '',
        '', '', 'python', '', 'Home', '',
        '', '', '', 'pandas', '',
        '', '', '', '']
values = [4000, 3181, 3386, 3055, 3467, 2244, 2898, 2484, 2112,
    1965, 1847, 1582, 1555, 1550, 1462, 1366, 1360, 1282, 1273, 1065]
# zip()
# list()
# *
datas = list(zip(names,values))
datas
```

```python
WordCloud().add(series_name="",data_pair=datas,word_size_range=[6,66]).set_global_opts().render_notebook()
```
