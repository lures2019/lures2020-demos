###### We can find that right columns are all have 2938 rows,but it's obsulute not true!
###### So we're supported to fix this data and remove some useless datas to make our predict module better!
###### The processing measure is to delete all the data of that row if it encounters an empty cell

```python
"""
    可以发现csv文件原先由2938行，但是部分列里面数据类型是Nan，所以需要剔除
    使用第三方库csv进行文件的读取和写入
"""
import csv

# 获取csv数据集文件所在的路径
path_now = os.path.join(DATASET_PATH,"Life Expectancy Data.csv")
f = open(path_now,mode="r",encoding="utf-8")
# 使用reader()方法来获取迭代对象
csv_reader = csv.reader(f)
# 通过列表解析式将所有的行内容集中到一个列表中
rows = [i for i in csv_reader]
# 获取csv文件的标头
headers = rows[0]
# 创建新的csv文件用来接收处理后的数据，其中mode="w+"表示追加写的方式创建文件，newline=""是csv文件写入时不自动换行,后面的编码格式是可以兼容中文
fp = open(DATASET_PATH + '/' + "data_end.csv",mode="w+",newline="",encoding="utf-8-sig")
# 使用writer()方法写入数据
csv_write = csv.writer(fp)
# 以行的方式写入标头
csv_write.writerow(headers)
# for循环是隔行处理原先的数据，删掉由Nan数据的行
for row in rows[1:]:
    flag = True
    for j in range(len(row)):
        if row[j].strip() == "":
            flag = False
            break
    if flag:
        csv_write.writerow(row)
fp.close()
```

### 1、The various predictors initially selected include demographic variables, income composition and mortality

###### Use multiple linear regression to explore the relationship between resource income composition and life expectancy

```python
# 导入机器学习常用的sklearn库
from sklearn.model_selection import train_test_split
# Linear regression
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
# 根据文件划分训练集和验证集，其中X是样本，包括4列数据，y是结果，只有预计寿命一个变量
X = data.loc[:,('Adult Mortality','infant deaths','under-five deaths ','Income composition of resources')]
y = data.loc[:,'Life expectancy ']
# 划分训练集和验证集，占比是8:2
X_train,X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2,random_state=100)
# 打印输出训练集和验证集各自的行数
print('X_train.shape={}\ny_train.shape ={}\nX_test.shape={}\ny_test.shape={}'.format(X_train.shape,y_train.shape, X_test.shape,y_test.shape))
# 首先使用简单线性回归来进行模拟训练，因为这不是一个(二)多分类问题，也不是判断问题，并且和时间关系也不是很大，所以决策树/随机森林/LSTM时间序列等模
# 所以使用的是多元线性回归来进行模拟训练
linreg = LinearRegression()
# 调用现成的fit()方法进行训练
model = linreg.fit(X_train, y_train)
print(model)
# Model intercept after training
print(linreg.intercept_)
# Model weight after training (no change in the number of features)
print(linreg.coef_)
```

```python
# 获取各个变量的系数
# y = -0.032*AM + 0.09 * id - 0.07 * uf + 23.72 * Icor
feature_cols = ['Adult Mortality','infant deaths','under-five deaths ','Income composition of resources']
B = list(zip(feature_cols,linreg.coef_))
print(B)
```

```python
# 用模型预测得出的预计寿命
y_pred = linreg.predict(X_test)
print(y_pred) 
```

```python
sum_mean = 0
for i in range(len(y_pred)):
    sum_mean+=(y_pred[i]-y_test.values[i])**2
sum_erro = np.sqrt(sum_mean/10)  
# 手动计算均方根误差
"""
    是一种常用的测量数值之间差异的量度，其数值常为模型预测的量或是被观察到的估计量。
    方均根偏移代表预测的值和观察到的值之差的样本标准差（sample standard deviation），
    当这些差值是以资料样本来估计时，他们通常被称为残差；当这些差值不以样本来计算时，
    通常被称为预测误差（prediction errors）。方均根偏移主要用来聚集预测里头误差的大小，通常是在不同的时间下，
    以一个量值来表现其预测的能力。方均根偏移是一个好的准度的量度，但因其与数值范围有关，
    因此被限制只能用来比较不同模型间某个特定变量的预知误差。
"""
print("RMSE by hand:",sum_erro)
# 绘制ROC曲线
"""
    是一种坐标图式的分析工具，用于 (1) 选择最佳的信号侦测模型、舍弃次佳的模型。 (2) 在同一模型中设定最佳阈值。
"""
plt.figure()
plt.plot(range(len(y_pred)),y_pred,'b',label="predict")
plt.plot(range(len(y_pred)),y_test,'r',label="test")
# 在画布上显示标签
plt.legend(loc="upper right") 
plt.xlabel("the number of test")
plt.ylabel('Life expectancy')
save_fig("linreg_plots")
plt.show()
```

### 9、Build a model to predict life

```python
X = data.loc[:,('Adult Mortality','infant deaths','Alcohol','percentage expenditure','Hepatitis B','Measles ','under-five deaths ','Polio','Total expenditure','Diphtheria ','GDP','Population','Income composition of resources')]
y = data.loc[:,'Life expectancy ']
X_train,X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2,random_state=100)
print('X_train.shape={}\ny_train.shape ={}\nX_test.shape={}\ny_test.shape={}'.format(X_train.shape,y_train.shape, X_test.shape,y_test.shape))
linreg = LinearRegression()
model = linreg.fit(X_train, y_train)
print(model)
# Model intercept after training
print(linreg.intercept_)
# Model weight after training (no change in the number of features)
print(linreg.coef_)
```

```python
feature_cols = ['Adult Mortality','infant deaths','Alcohol','percentage expenditure','Hepatitis B','Measles ','under-five deaths ','Polio','Total expenditure','Diphtheria ','GDP','Population','Income composition of resources']
B = list(zip(feature_cols,linreg.coef_))
print(B)
```

```python
# predict
y_pred = linreg.predict(X_test)
print(y_pred) 
RMSE = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
print('RMSE:',RMSE)
```

```python
# Calculated error rate
# Make ROC curve
plt.figure()
plt.plot(range(len(y_pred)),y_pred,'b',label="predict")
plt.plot(range(len(y_pred)),y_test,'r',label="test")
# Show labels in the figure
plt.legend(loc="upper right") 
plt.xlabel("the number of test")
plt.ylabel('Life expectancy')
save_fig("end_predict_plots")
plt.show()

plt.scatter(y_test, y_pred)
plt.plot([y_test.min(),y_test.max()], [y_test.min(),y_test.max()], 'k--')
plt.xlabel('Test')
plt.ylabel('Predict')
```

```python
# 可以看到使用多元线性回归进行预测分析时，x轴为90处点的偏离程度较大，于是剔除该部分的点进行模型的优化
drop_index = data[data['Life expectancy ']== 90].index.values
X = X.drop(drop_index)
y = y.drop(drop_index)
# The parameter random_state that divides the training set and the test set here are both 1, indicating that the randomly allocated data is the same group
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
# 对训练集进行训练
lr = LinearRegression()
lr.fit(X_train, y_train)
# 对测试集进行预测
y_pred = lr.predict(X_test)
RMSE = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
# 可以看出通过模型的优化后误差减小了
print('RMSE:',RMSE)
```
