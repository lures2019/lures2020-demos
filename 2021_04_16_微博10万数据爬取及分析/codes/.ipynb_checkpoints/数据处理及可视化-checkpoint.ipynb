```python
import pandas as pd
%matplotlib inline
from matplotlib import pyplot as plt
import requests
from pyecharts import options as opts
from pyecharts.charts import Scatter
import os

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
path = 'pictures'
if not os.path.exists(path):
    os.mkdir(path)
```

### 1、分布散点图

```python
# 选取所在地区不是其他或者NaN类型的用户，使用高德api进行逆编码，再绘制地图
# 删除[所在地区]中值为'其他'的项
area = df.loc[df['所在地区'] != '其他']
# 去除缺失值数据
area = area.dropna(axis = 0)
# 将所在的地区设置成列表形式
locations = list(area['所在地区'])
```

```python
# 对locations中位置进行遍历，以减少逆编码工作量
my_dict = {}
for location in locations:
    if location not in my_dict:
        my_dict[location] = 1
    else:
        my_dict[location] += 1
locations = list(my_dict.keys())
values = list(my_dict.values())

# 对于有空格的locations需要删去空格,否则逆编码出现问题
locations = [i.replace(" ","") for i in locations]
print(locations,values)
print(len(locations),len(values))
```

```python
%%time
# 进行地理编码
latitude = []
longtitude = []
# 存储正确信息的列表
new_locals = []
new_values = []
for i in range(len(locations)):
    url = 'https://restapi.amap.com/v3/geocode/geo?address={}&output=json&key=84bb08664d7f042079e48b576d9dd7c1'.format(locations[i])
    response = requests.get(url)
    response.encoding = response.apparent_encoding
    try:
        data = response.json()['geocodes'][0]['location'].split(',')
        latitude.append(float(data[0]))
        longtitude.append(float(data[1]))
        new_locals.append(locations[i])
        new_values.append(values[i])
    except Exception as error:
        print(locations[i],error)
```

```python
# 将经纬度等信息导出为csv文件
columns = ['所在地区','用户数量','经纬度']
# '85.746345,48.37248'
my_list = [[] for i in range(len(latitude))]
for i in range(len(latitude)):
    my_list[i].append(new_locals[i])
    my_list[i].append(new_values[i])
    my_list[i].append('{},{}'.format(latitude[i],longtitude[i]))

test = pd.DataFrame(columns=columns,data=my_list)
test.to_csv(path + '/' + "地图文件.csv",encoding='utf-8-sig',index=None)
```

```python
# 设置了图片大小，和图片的dpi
fig = plt.figure(figsize=(20, 8), dpi=80)
plt.scatter(latitude, longtitude, marker = '*',color = 'dodgerblue')
plt.xlabel("经度")
plt.ylabel("纬度")
plt.savefig(path + '/' + '用户分布散点图.png')
plt.show()
```

```python
# 各时段转发微博数量面积图
from pyecharts import options as opts
from pyecharts.charts import Line
import pyecharts

pyecharts.globals._WarningControl.ShowWarning = False
forwards = list(df['转发时间'])
# 因为范围过大，就根据天来划分了
time_dict = {}
for i in forwards:
    value = i.split(" ")[0]
    if value == '利':
        pass
    else:
        if value not in time_dict:
            time_dict[value] = 1
        else:
            time_dict[value] += 1
# 获取时间和数值
days = list(time_dict.keys())
values = list(time_dict.values())
a = Line()
a.add_xaxis(xaxis_data=days)
a.add_yaxis(
    series_name="",
    y_axis=values,
    symbol="emptyCircle",
    is_symbol_show=True,
    label_opts=opts.LabelOpts(is_show=False),
    #设置面积图填充颜色
    areastyle_opts=opts.AreaStyleOpts(opacity=1, color="green"),
)
a.set_global_opts(
    tooltip_opts=opts.TooltipOpts(is_show=False),
    yaxis_opts=opts.AxisOpts(
        type_="value",
        name = '用户数',
        axistick_opts=opts.AxisTickOpts(is_show=True),
        splitline_opts=opts.SplitLineOpts(is_show=True),
    ), 
    xaxis_opts=opts.AxisOpts(type_="category",name = '时间',boundary_gap=False),
)
# 设置 boundary_gap 的时候一定要放在最后一个配置项里, 不然会被覆盖
a.render(path + '/' + '各时段转发微博数量面积图.html')
a.render_notebook()
```

```python
# 用户性别占比饼图
from pyecharts.charts import Pie
from pyecharts import options

sexes = list(df['性别'])
boys = 0
girls = 0
for sex in sexes:
    if sex == 'f':
        girls += 1
    elif sex == 'm':
        boys += 1
    else:
        pass
# 绘制男女占比图
data1 = [("男生", boys), ("女生", girls)]
pie = Pie()
pie.add(
    # 设置系列名称
    series_name="男女转发占比图",
    # 设置需要展示的数据
    data_pair=data1,
    # 设置圆环空心部分和数据显示部分的比例
    radius=["30%", "70%"],
    # 设置饼是不规则的
    rosetype="radius"

)
pie.set_series_opts(label_opts=options.LabelOpts(formatter="{b}: {d}%"))
pie.set_global_opts(title_opts=options.TitleOpts(title="男女转发占比图"))
pie.render(path + '/' + '男女转发占比图.html')
pie.render_notebook()
```

```python
# 用户的年龄段占比饼图
ages = df['出生日期']
# 去除缺失值数据
ages = ages.dropna(axis = 0)
# 选择符合条件的数据,微博以60后开始统计，到10前
ages_dict = {
    '60后':0,
    '70后':0,
    '80后':0,
    '90后':0,
    '00后':0,
    '10后':0
}
for age in ages:
    try:
        year = age.split('-')[0]
        if len(year) == 4:
            year = int(year)
            if year >= 1960 and year < 1970:
                ages_dict['60后'] += 1
            elif year < 1980:
                ages_dict['70后'] += 1
            elif year < 1990:
                ages_dict['80后'] += 1
            elif year < 2000:
                ages_dict['90后'] += 1
            elif year < 2010:
                ages_dict['00后'] += 1
            elif year < 2020:
                ages_dict['10后'] += 1
            else:
                pass
    except Exception as e:
        pass
# 绘制年龄占比图
data1 = [("60后", ages_dict['60后']), 
         ("70后", ages_dict['70后']),
         ("80后", ages_dict['80后']),
         ("90后", ages_dict['90后']),
         ("00后", ages_dict['00后']),
         ("10后", ages_dict['10后'])
        ]
pie = Pie()
pie.add(
    # 设置系列名称
    series_name = "转发用户年龄占比图",
    # 设置需要展示的数据
    data_pair = data1,
    # 设置圆环空心部分和数据显示部分的比例
    radius = ["30%", "70%"],
    # 设置饼是不规则的
    rosetype = "radius"

)
pie.set_series_opts(label_opts = options.LabelOpts(formatter="{b}: {d}%"))
pie.set_global_opts(title_opts = options.TitleOpts(title="转发用户年龄占比图"))
pie.render(path + '/' + '转发用户年龄占比图.html')
pie.render_notebook()
```
